# EDA
Exploratory Data Analysis (EDA) is a crucial step in understanding your dataset before applying machine learning models.

Exploratory Data Analysis (EDA) is a crucial step in understanding your dataset before applying machine learning models.

 It involves:
1. <b>Understanding Data Structure</b> – Checking data types, dimensions, and missing values.
2. <b>Summarizing Data</b> – Using statistical measures like mean, median, and standard deviation.
3. <b>Visualizing Data</b> – Using plots (histograms, boxplots, scatterplots) to identify trends and relationships.
4. <b>Detecting Outliers and Anomalies</b> – Finding extreme values that may affect analysis.
5. <b>Handling Missing Values</b> – Deciding whether to impute, drop, or ignore missing data.
6. <b>Feature Engineering</b> – Creating new features or modifying existing ones for better model performance.

# Skewness
 When data points on a bell curve are not distributed symmetrically to the left and right sides of the median, the bell curve is skewed.
 
 OR
 
 skewness describes how much a dataset's distribution is asymmetrical or lopsided, meaning the data isn't evenly spread around the center. If the tail (the end) is longer on one side, it's skewed, either to the left or right. 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic.csv")
```

```python
df.head(4)
```

# 1. Categorical Data

   * countplot

```python
sns.countplot(x=df["Survived"])  # Use x= instead of passing the column directly
plt.title("Survival Count")
plt.show()
df['Survived'].value_counts()
```

```python
sns.countplot(x=df["Pclass"]) 
```

#  Piechart 

```python
df['Survived'].value_counts().plot(kind="pie", autopct="%.2f")
```
