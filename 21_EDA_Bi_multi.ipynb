# EDA using Bivarient, Multivarient

```python
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
df = pd.read_csv("Titanic.csv")
df.head(3)
```

```python
flights=sns.load_dataset("flights")
flights.head()
```

```python
iris= sns.load_dataset("iris")
iris.head()
```

1. scatterplon(Numerical - Numerical)

```python
sns.scatterplot(x=tips["total_bill"], y=tips['tip']) # Bivarient
```

```python
sns.scatterplot(x=tips["total_bill"], y=tips['tip'],hue=tips["sex"]) # Multivarient
```

```python
sns.scatterplot(x=tips["total_bill"], y=tips['tip'],hue=tips["sex"],style=tips["smoker"], size=tips["size"])
```

2. Bar plot(Numerical - Categorical)

```python
sns.barplot(x=df["Pclass"], y=df["Age"])
```

3. BoxPlot(Numerical - catagories)

```python
sns.boxplot(x=df["Sex"], y=df["Age"],hue=df["Survived"])
```

4. Distplot(Numerical - catagories)

```python
sns.distplot(df[df["Survived"]==0]["Age"],hist=False)
sns.distplot(df[df["Survived"]==1]["Age"],hist=False)
```

5. HeatMap(catagories - catagories)

```python
sns.heatmap(pd.crosstab(df["Pclass"], df["Survived"]))
```

6. ClusterMap(catagories - catagories)

```python
sns.clustermap(pd.crosstab(df["SibSp"], df["Survived"]))
```

7. Pearplot()

```python
sns.pairplot(iris)
```
